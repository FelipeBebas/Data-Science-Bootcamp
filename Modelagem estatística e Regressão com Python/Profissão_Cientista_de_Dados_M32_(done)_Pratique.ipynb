# Probabilidade - Tarefa

```python
import numpy as np
from scipy import stats
```

1. Uma máquina embala ração para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatório, com distribuição Normal, com média 15kg, mas com desvio padrão 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produção tenha mais que 15,2 kg.

```python
# Definir os parâmetros da distribuição Normal
mu = 15  # média
sigma = 0.1  # desvio padrão

# Definir o valor para o qual queremos calcular a probabilidade
x = 15.2

# Calcular a probabilidade de um pacote ter mais que 15.2 kg
probabilidade = stats.norm.sf(x, loc=mu, scale=sigma)

print(f"A probabilidade de um pacote ter mais que {x} kg é: {probabilidade:.4f}")
```

2. Da máquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Definir o percentil desejado
percentil_ordem = 0.95

# Calcular o valor do peso para o percentil de ordem 95%
valor_percentil = stats.norm.ppf(percentil_ordem, loc=mu, scale=sigma)

print(f"O percentil de ordem {percentil_ordem*100:.0f}% para o peso dos pacotes é: {valor_percentil:.4f} kg")
```

3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na célula abaixo:

```python
Y=np.linspace(-6, 6, 200)
# Seu código começa aqui
prob_Y1 = stats.norm.cdf(Y, loc=0, scale=1)

# Calcular a probabilidade de Y2 ser menor que cada valor em Y (t-student com 5 graus de liberdade)
df = 5 # graus de liberdade
prob_Y2 = stats.t.cdf(Y, df=df)

print("Probabilidades para Y1 (Normal Padrão):")
print(prob_Y1)
print("\nProbabilidades para Y2 (t-student com 5 graus de liberdade):")
print(prob_Y2)
```

4. Faça uma visualização gráfica dos dados do item anterior.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.lineplot(x=Y, y=prob_Y1, label='Normal Padrão')
sns.lineplot(x=Y, y=prob_Y2, label='t-student (5 Graus de Liberdade)')

plt.title('Comparação das CDFs: Normal Padrão vs. t-student')
plt.xlabel('Valores de Y')
plt.ylabel('Probabilidade Acumulada P(X < Y)')
plt.legend()
plt.grid(True)
plt.show()
```

5. 3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
Y=np.linspace(0, 1, 20)
# Seu código começa aqui

# Definir os graus de liberdade para a distribuição t-student
df_t = 5

# Calcular o percentil para a distribuição Normal Padrão
percentil_Y1 = stats.norm.ppf(Y, loc=0, scale=1)

# Calcular o percentil para a distribuição t-student
percentil_Y2 = stats.t.ppf(Y, df=df_t)

print("Percentis para Y1 (Normal Padrão):")
print(percentil_Y1)
print("\nPercentis para Y2 (t-student com 5 graus de liberdade):")
print(percentil_Y2)
```

6. Faça um gráfico com os dados do item anterior.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.lineplot(x=Y, y=percentil_Y1, label='Normal Padrão')
sns.lineplot(x=Y, y=percentil_Y2, label='t-student (5 Graus de Liberdade)')

plt.title('Comparação dos Percentis: Normal Padrão vs. t-student')
plt.xlabel('Probabilidade Acumulada (Y)')
plt.ylabel('Valor do Percentil')
plt.legend()
plt.grid(True)
plt.show()
```
